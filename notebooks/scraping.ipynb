```python
import requests
import json

# Базовый URL для запросов
base_url = "https://api.nhle.com/stats/rest/en/team/summary"

# Параметры запроса
query_params = {
    "isAggregate": "false",
    "isGame": "true",
    "sort": '[{"property":"gameDate","direction":"ASC"},{"property":"teamId","direction":"ASC"}]',
    "start": 0,
    "limit": 100,
    "cayenneExp": "gameTypeId=2 and seasonId<=20232024 and seasonId>=20032004",
}

# Заголовки запроса
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:128.0) Gecko/20100101 Firefox/128.0",
    "Accept": "*/*",
    "Accept-Language": "ru-RU,ru;q=0.8,en-US;q=0.5,en;q=0.3",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Referer": "https://www.nhl.com/",
    "Origin": "https://www.nhl.com",
    "Sec-Fetch-Dest": "empty",
    "Sec-Fetch-Mode": "cors",
    "Sec-Fetch-Site": "cross-site",
    "Connection": "keep-alive",
    "Priority": "u=4",
    "TE": "trailers",
}

all_games = []  # Список для всех игр
total_games = 10000  # Примерное количество всех игр
batch_size = 100  # Размер пакета

# Цикл для получения всех игр
for start in range(0, total_games, batch_size):
    query_params["start"] = start
    response = requests.get(base_url, headers=headers, params=query_params)
    data = response.json()
    games = data.get("data", [])
    all_games.extend(games)
    if len(games) < batch_size:
        break  # Выход из цикла, если меньше 100 игр в пакете (конец данных)

print(f"Total games fetched: {len(all_games)}")
```

```python
with open("nhl_games.json", "w") as f:
    json.dump(all_games, f, indent=4)
```

```python
import requests
import pandas as pd
import json
```

```python
# Базовый URL для запросов
base_url = "https://api.nhle.com/stats/rest/en/team/summary"

# Параметры запроса для тестирования
query_params = {
    "isAggregate": "false",
    "isGame": "true",
    "sort": '[{"property":"gameDate","direction":"ASC"},{"property":"teamId","direction":"ASC"}]',
    "limit": 100,
    "start": 0,
    "cayenneExp": "gameTypeId=2 and seasonId=20232024",  # Пример одного сезона
}

# Заголовки запроса
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:128.0) Gecko/20100101 Firefox/128.0",
    "Accept": "*/*",
    "Accept-Language": "ru-RU,ru;q=0.8,en-US;q=0.5,en;q=0.3",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Referer": "https://www.nhl.com/",
    "Origin": "https://www.nhl.com",
    "Sec-Fetch-Dest": "empty",
    "Sec-Fetch-Mode": "cors",
    "Sec-Fetch-Site": "cross-site",
    "Connection": "keep-alive",
    "Priority": "u=4",
    "TE": "trailers",
}


def fetch_data(season, start):
    """Функция для получения данных с API"""
    query_params["cayenneExp"] = f"gameTypeId=2 and seasonId={season}"
    query_params["start"] = start
    try:
        response = requests.get(base_url, headers=headers, params=query_params)
        response.raise_for_status()  # Проверка на успешность запроса
        return response.json()
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e}")
    except requests.exceptions.SSLError as e:
        print(f"SSL Error: {e}")
    except requests.exceptions.RequestException as e:
        print(f"Request Exception: {e}")
    return None


# Пример одного сезона для тестирования
season = 20232024
start = 0

# Попробуем получить данные за один сезон без пагинации
all_games = []
while True:
    data = fetch_data(season, start)
    if not data or "data" not in data or not data["data"]:
        break
    games = data.get("data", [])
    all_games.extend(games)
    start += len(games)
    print(f"Fetched {len(games)} games starting at index {start - len(games)}")

print(f"Total games fetched: {len(all_games)}")

# Вывод полного ответа для отладки
if all_games:
    print(json.dumps(all_games[0], indent=4))

# Создание DataFrame и сохранение данных в CSV
df = pd.DataFrame(all_games)
print(df.head())

# Установка порядка столбцов
column_order = [
    "gameId",
    "gameDate",
    "teamId",
    "teamFullName",
    "homeRoad",
    "opponentTeamAbbrev",
    "goalsFor",
    "goalsAgainst",
    "goalsForPerGame",
    "goalsAgainstPerGame",
    "shotsForPerGame",
    "shotsAgainstPerGame",
    "penaltyKillPct",
    "penaltyKillNetPct",
    "powerPlayPct",
    "powerPlayNetPct",
    "faceoffWinPct",
    "pointPct",
    "wins",
    "losses",
    "ties",
    "winsInRegulation",
    "winsInShootout",
    "regulationAndOtWins",
    "otLosses",
    "points",
    "gamesPlayed",
]

# Проверка наличия всех столбцов и заполнение отсутствующих значений
missing_columns = [col for col in column_order if col not in df.columns]
if missing_columns:
    print(f"Missing columns: {missing_columns}")
else:
    df = df.fillna(
        {
            "faceoffWinPct": 0.0,
            "penaltyKillPct": 0.0,
            "penaltyKillNetPct": 0.0,
            "powerPlayPct": 0.0,
            "powerPlayNetPct": 0.0,
            "shotsAgainstPerGame": 0.0,
            "shotsForPerGame": 0.0,
        }
    )

    df = df[column_order]
    df.to_csv("nhl_all_games2324.csv", index=False)
```
