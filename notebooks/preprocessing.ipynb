```python
# Перемещаем столбцы в нужном порядке
column_order = [
    "gameId",
    "gameDate",
    "teamFullNameEncoded",
    "opponentTeamAbbrevEncoded",
    "homeRoadEncoded",
    "gamesPlayed",  # Соответствует GP
    "wins",  # Соответствует W
    "losses",  # Соответствует L
    "ties",  # Соответствует T
    "otLosses",  # Соответствует OT
    "points",  # Соответствует P
    "pointPct",  # Соответствует P%
    "winsInRegulation",  # Соответствует RW
    "regulationAndOtWins",  # Соответствует ROW
    "winsInShootout",  # Соответствует S/O Win
    "goalsFor",  # Соответствует GF
    "goalsAgainst",  # Соответствует GA
    "goalsForPerGame",  # Соответствует GF/GP
    "goalsAgainstPerGame",  # Соответствует GA/GP
    "powerPlayPct",  # Соответствует PP%
    "penaltyKillPct",  # Соответствует PK%
    "powerPlayNetPct",  # Соответствует Net PP%
    "penaltyKillNetPct",  # Соответствует Net PK%
    "shotsForPerGame",  # Соответствует Shots/GP
    "shotsAgainstPerGame",  # Соответствует SA/GP
    "faceoffWinPct",  # Соответствует FOW%
    "teamId",  # ID команды
    "teamIdNumeric",  # Numeric ID команды
]

# Переставляем столбцы в нужном порядке
df = df[column_order]

# Проверка результата
print(df.head())
```

```python
df["powerPlayPct"] = df["powerPlayPct"].round(2)
df["penaltyKillPct"] = df["penaltyKillPct"].round(2)
df["powerPlayNetPct"] = df["powerPlayNetPct"].round(2)
df["powerPlayNetPct"] = df["powerPlayNetPct"].round(2)
```

```python
df["faceoffWinPct"] = df["faceoffWinPct"].round(2)
```

```python
df = df.drop(columns=["teamId", 'teamIdNumeric'])
```
